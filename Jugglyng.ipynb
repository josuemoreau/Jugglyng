```python
import modele
import ipywidgets
import pythreejs

model = modele.Model([{"color":"blue", "tone":"do"}, {"color":"red","tone":"mi"}], ["green", "purple"],
                    pattern=[(5, 1), (3, 0), (4, 0), (5, 0), (3, 1), (4, 1)])
view = modele.View(model)

slider = ipywidgets.FloatSlider(min=0, max=40)
slider.observe(lambda change: view.update(slider.value), names="value")
ipywidgets.VBox([view.widget,slider])
```

```python
def summary(state):
    for b in state.balls:
        if b.time_to_land:
            print("{}:\t {}-{}-{}->{}".format(b.color, b.source_hand, b.time_flying, b.time_to_land, b.target_hand))
        else:
            print("{}:\t {}".format(b.color, b.target_hand))

import pprint
for state in model.states: 
    summary(state)
    print()
```

```python
import pdb
pdb.pm()
state = model.state(0)
throw = modele.Throw(0,1,3)
pdb.run("model.transition(state, throw)")
```
